```python
x = 10 #Global variable

def update_x():
    global x #Declare that we are using the global variable x
    x += 5  #modify the global variable

update_x()
print(x) #output:15
```

```python
x = 10 #Global variable

def update_x():
    globals()['x'] += 5  #modify the global variable

update_x()
print(x) #output:15
```

```python
def count(lst):
    even = 0
    odd = 0

    for i in lst:
        if i%2 == 0:
            even += 1
        else:
            odd +=1
    return even,odd
lst = [10, 9, 8, 23, 50]
even, odd =  count(lst)

print(even)
print(odd)
```

```python
#FIBONACCI SEQUESNCE

def fib(n):
    a=0
    b=1

    print(a)
    print(b)

    for i in range(0, n):
        c = a+b
        a = b
        b = c

        print(c)

fib(10)
    
```

```python
#FACTORIALS

def fact(n):
    if n==0:
        return 1
    return n * fact(n-1)

result = fact(5)

result
```
