```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
tpl = tuple(mpg['cylinders'])
sns.displot(tpl, bins=range(3,10,1), kde= True)
```

```python
mpg['origin'].unique()
```

```python
sns.countplot(x = 'cylinders', data = mpg, color = 'purple', hue = 'origin')
plt.legend(loc = 'lower left')
```

```python
sns.boxplot(x='origin', y='acceleration', data=mpg)
```

```python
sns.jointplot(x='horsepower', y='acceleration', data=mpg)
```

```python
#subplot
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.scatterplot(x='horsepower', y= 'acceleration' , data=mpg, hue='origin')
plt.subplot(2,2,2)
sns.scatterplot(x='horsepower', y='acceleration', alpha=0.5, data=mpg)
plt.subplot(2,2,3)
sns.boxplot(x='acceleration', y='origin', data=mpg)
plt.savefig('d:\py\myplot.png')
```

```python
sns.lmplot(x='horsepower', y='acceleration', data=mpg, hue='origin')
```

```python
sns.pairplot(iris, hue='species', palette= 'BrBG', markers=['o','d','s'])
```
