```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Admin/Downloads/matplotlib/matplotlib/score.csv')
```

```python
data.head()
```

```python
x = data['marks']
y = data['rank']
plt.plot(x,y, color='green')
plt.scatter(x,y, color='orange')
plt.xlabel("Marks", color='blue')
plt.ylabel("Rank", color='blue')
plt.title("EAMCET EXAMS", color= 'red')
plt.show()
```

```python
#BAR GRAPH

plt.bar(x,y, color='red')
plt.xlabel("Marks", color='blue')
plt.ylabel("Rank", color='blue')
plt.title("EAMCET EXAMS", color= 'red')
plt.xlim(800, 850)
plt.ylim(1500,2100)
plt.show()
```
