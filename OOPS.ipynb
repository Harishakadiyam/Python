```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Call parent constructor
        self.breed = breed

    def speak(self):
        print(f"{self.name} ({self.breed}) barks")

d = Dog("Tommy", "Beagle")

d.speak()  # Output: Tommy (Beagle) barks
```

```python
#ABSTRACT METHOD
from abc import ABC, abstractmethod

class computer(ABC): # This is special class from ABC module it will allow you to create Abstract class
    @abstractmethod
    def process(self): #declaration
        pass #since process has no body pass


class Laptop(computer): #Laptop is subclass of computer, computer is a ABC method class
    def process(self): #declaration ( it overrides the process of abstract method
        print('its working') #definition
     

com1 = Laptop() #object 
#com = computer()
com1.process() #calling
```

```python
from time import sleep
from threading import *

class hello(Thread):
    def run(self):
        for i in range(10):
            print('hello')
            sleep(1)
            
class hi(Thread):
    def run(self):
        for i in range(10):
            print('hi')
            sleep(1)

t1 = hello() #t1 is the object of hello method
t2 = hi()   # t2 is the objec of hi method

t1.start()
t2.start()
```
